```python
#7th May, 2026
#Data Cleaning
#Renaming using ".rename" and ",inplace=True"
df.rename(columns={
    "EmpId": "emp_id",
    "fullName": "full_name",
    "Dept": "department",
    "Age": "age",
    "Salary($)": "salary",
    "JoinDate": "join_date",
    "Performance": "performance",
    "Remote?": "remote"
},inplace=True)

df
```

```python
#Columns Transformation using 
df['full_name'] = (
    df['full_name']
    .str.strip()
    .str.title()
    .str.replace(r"[\s-]+", " ", regex=True)
)

df['full_name'].to_list()

df
```

```python
#Replacing inconsistent values
df.replace({
    'hr': 'HR',
    'Mktng': 'Marketing',
    'marketing': 'Marketing',
    None: np.nan
}, inplace=True)

df['department']
```

```python
#Cleaning the Age Column
#def defines function
#isinstance(value) helps confirm the data type
#isdigit returns numbers that are in string format
#you must define the function "df['age'] = "

def clean_age(value):
    if value == "thirty-five":
        return 25
    elif isinstance(value, (int,float)):
        return (value)
    elif isinstance(value, str) and value.isdigit():
        return int(value)
    else:
        return np.nan
df['age'] = df['age'].apply(clean_age)
df['age']
        
```

```python
#Clean Salary

df['salary'] = (
    df['salary']
    .astype(str)
    .replace(r"[$,]", "", regex=True)
    .replace("NaN", np.nan)
    .astype(float)
)

df['salary']
```

```python
#Clean Perfomance
df['performance'] = (
    df['performance']
    .astype(str)
    .replace(r"%", " ", regex=True)
    .replace("None", np.nan)
    .astype(float)
)
df['performance']
```

```python
#Cleaning remote

df.replace({
    "yes": "Yes",
    "no": "No",
    "N": "No",
    "Y": "Yes",
    "Y": "Yes",
    "True": "Yes",
    None: np.nan,
}, inplace=True)

df
    
```

```python
#Cleaning inconsistent date value using pd.to_datetime(df['join_date'] 
#pd.to_datetime(df['join_date'] cannot handle complex date format, so we need to import "parser" from dateutil

#df['join_date'] = pd.to_datetime(df['join_date'], errors="coerce")
#df['join_date']

def date_parser(date_str):
    try:
        return parser.parse(date_str, dayfirst=True)
    except Exception:
        return pd.NaT

df['join_date'] = df['join_date'].apply(date_parser)
df['join_date']
```

```python
#Changing number in string format using ".to_numeric"
df['emp_id'] = pd.to_numeric(df['emp_id'], errors='coerce')
df['emp_id']
```

```python
df.info()
```

```python
#Handling missing values using ".fillna"
df['full_name'] = df['full_name'].fillna('Unknown')
df['full_name']
```

```python
#Replacing the missing value in department with mode

df['department'] = df['department'].fillna(df['department'].mode()[0])
df['department']
```

```python
#Replacing the missing value in age with the median
df['age'] = df['age'].fillna(df['age'].median())
df['age']
```

```python
#Replacing the missing value in salary with the median

df['salary'] = df['salary'].fillna(df['salary'].median())
df['salary']
```

```python
#Replacing the missing value in performance with the median
df['performance'] = df['performance'].fillna(df['performance'].median())
df['performance']
```

```python
df['remote'] = df['remote'].fillna(df['remote'].mode()[0])
df['remote']
```

```python
#Analysing Data using 'df.groupby()'
#FINDING AVERAGE OF SALARY USING DEPARTMENT
#.sort_values(ascending=False) SORTING BY SALARY


merge_df.groupby('department')['salary'].mean().sort_values(ascending=False)
```

```python
#FINDING AVERAGE OF SALARY AND PERFORMANCE USING DEPARTMENT
#.sort_values(by=['performance', 'salary'], ascending=[False, False]) SORTING MULTIPLE
#.sort_values(by=['salary'], ascending=[False]) SORTING BY SALARY

merge_df.groupby('department')[['performance', 'salary']].mean().sort_values(by=['performance', 'salary'], ascending=[False, False])
```

```python
#18th May
#Writing multiple aggregations at once using '.agg'

merge_df.groupby('department').agg(
    avg_salary=('salary', 'mean'),
    total_salary=('salary', 'sum'),
    total_employee=('salary', 'count'),
    min_salary=('salary', 'min'),
    max_salary=('salary', 'max'),
)
```

```python
#Writing multiple aggregations at once using '.agg' and multiple fields

merge_df.groupby('department').agg(
    avg_salary=('salary', 'mean'),
    avg_performance=('performance', 'mean'),
    total_employee=('emp_id', 'count')
).reset_index()
```

```python
#Finding Range using Lambda
merge_df.groupby('department')['salary'].apply(lambda x: x.max() - x.min())
```

```python
#Filtering using lambda
merge_df.groupby('department').filter(lambda x: x['salary'].mean() >= 5000)
```

```python
#Adding a new column using transform

merge_df['department_avg_salary'] = merge_df.groupby('department')['salary'].transform('mean')
merge_df
```

```python
#Detecting outliers using the ".quantile()"
Q1 = merge_df['salary'].quantile(0.25)
Q3 = merge_df['salary'].quantile(0.75)

IQR = Q3 - Q1

IQR
```

```python
#Finding the Lower and upper bound
lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR

outliers = (merge_df['salary'] < lower_bound) | (merge_df['salary'] > upper_bound)

print('Outliers:', outliers.sum())

lower_bound
upper_bound
```

```python
#Excluding Outliers from dataframe using ""

lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR

new_df = merge_df[
    (merge_df['salary'] > lower_bound) & (merge_df['salary'] < upper_bound)
]

new_df
```

```python
#VISUALIZATION USING MATPLOTLIB 
#Average salary by department
#declare "df_sorted" to sort in descending order


df_sorted = merge_df.sort_values('department_avg_salary', ascending=False)
plt.bar(df_sorted['department'], df_sorted['department_avg_salary'])
plt.title('AVERAGE SALARY BY DEPARTMENT')
plt.xlabel('Department')
plt.ylabel('Salary')
plt.grid(axis='y', linestyle='-', alpha=1)

```
